```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sin_wave = np.sin([i*np.pi for i in range(100)])
```

```python
open_ = 2901974
close_ = 2880811
(close_-open_)/open_
```

```python
plt.plot(sin_wave)
```

```python
fig, ax = plt.subplots(figsize=(10, 6), facecolor="w")
ax.plot(sin_wave,label='label',marker='o',markersize=7)
```

```python
plt.clf()

ax = plt.gca()
ax.set_xlabel('Day')
ax.set_ylabel('Stock price')
ax.set_xticks(range(0,20,2))
ax.plot(sin_wave)
```
